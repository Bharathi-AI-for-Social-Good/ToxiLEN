```python
import pandas as pd
from torch import nn
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
```

```python
device = 'cuda'
model_name = 'Salesforce/blip-image-captioning-base'
```

```python
train_df = pd.read_csv('D:\ChineseMemesMisogynyAnalysis\data\\all\\test.csv')
file_names = train_df['filename'].to_list()
```

```python
processor =  BlipProcessor.from_pretrained(model_name)
model = BlipForConditionalGeneration.from_pretrained(model_name).to(device)
```

```python
def generate_caption(image_path):
    raw_image = Image.open(image_path).convert('RGB')
    inputs = processor(raw_image,return_tensors='pt').to(device)
    out = model.generate(**inputs)
    caption = processor.decode(out[0], skip_special_tokens=True)
    
    return caption
```

```python
caption_outputs = []
for image_path in file_names:
    image = rf"D:\ChineseMemesMisogynyAnalysis\data\images\test\{image_path}"
    caption = generate_caption(image_path=image)    
    caption_outputs.append(caption)
    print(caption)
```

```python
train_df['captions'] = caption_outputs
train_df.to_csv('test_caption_version.csv')
```
